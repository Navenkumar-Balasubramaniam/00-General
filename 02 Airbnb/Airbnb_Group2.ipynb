Q16. Examine the "license" field a bit more closely. There are different structures present, apart from the one described above. Try to identify them and count how many listings have each type of license.

```python
df = airbnb_listings_df.copy()

# Create an empty column for categories
df["license_category"] = "Other"

# 1. Missing licenses
df.loc[df["license"].isna(), "license_category"] = "Missing"

# Convert everything to string once (makes checks easier)
lic = df["license"].astype(str)

# 2. Long ES codes (start with ES)
df.loc[(df["license_category"] == "Other") & (lic.str.startswith("ES") | lic.str.startswith("es") ), "license_category"] = "Long ES codes"

# 3. Short VT codes (start with VT-)
df.loc[(df["license_category"] == "Other") & (lic.str.startswith("VT") | lic.str.startswith("vt") ), "license_category"] = "VT codes"

# 4. Short AM codes (start with AM-)
df.loc[(df["license_category"] == "Other") & (lic.str.startswith("AM") | lic.str.startswith("am")), "license_category"] = "AM codes"

# 5. Short AM codes (start with HM-)
df.loc[(df["license_category"] == "Other") & (lic.str.startswith("HM") | lic.str.startswith("hm")) , "license_category"] = "HM codes"

# 6. Fraction codes (contain "/")
df.loc[(df["license_category"] == "Other") & lic.str.contains("/", na=False), "license_category"] = "Fraction codes"

# 7. Numeric-only codes (only digits)
df.loc[(df["license_category"] == "Other") & lic.str.isdigit(), "license_category"] = "Numeric-only codes"

# 8. Verbal descriptions (letters only, like “Exempt”, “En proceso”)
df.loc[
    (df["license_category"] == "Other") & lic.str.isalpha(),
    "license_category"
] = "Verbal descriptions"

# Count how many listings fall in each category
category_counts = df["license_category"].value_counts()

print(category_counts)
```
